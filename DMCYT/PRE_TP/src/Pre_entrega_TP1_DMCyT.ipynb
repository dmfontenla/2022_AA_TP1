#Recorte circular

```python
first_img = df_flowers.loc[5]['image']
import numpy as np
copy_img = first_img.copy()
 
rows = 128
cols = 128
size = rows*cols
 
A = np.array([0]*size).reshape(rows,cols)
 
m1 = 64
m2 = 64
radiusSquared = 3000

print(A.shape)
print(f'matrix with dimension {rows} x {cols} is {A}')
print(A)
for y in range(128):
  for x in range(128):
    r = copy_img[x,y][0]
    g = copy_img[x,y][1]
    b = copy_img[x,y][2]
    #print(x)

    dx = x - m1;
    dy = y - m2;
    distanceSquared = dx * dx + dy * dy;

    if (distanceSquared >= radiusSquared) :
        copy_img[x][y][0] = 255
        copy_img[x][y][1] = 255
        copy_img[x][y][2] = 255

plt.imshow(copy_img,cmap='gray')
plt.axis('off')
```

#Brillo

```python
first_img = df_flowers.loc[0]['image']
copy_img = first_img.copy()

import numpy as np
 
rows = 128
cols = 128
size = rows*cols
 
A = np.array([0]*size).reshape(rows,cols)
 
print(A.shape)
print(f'matrix with dimension {rows} x {cols} is {A}')
print(A)

def truncate(x):
    '''makes sure returned value is between 0 and 255'''
    return min(255, max(0, x))

d = 25

for y in range(128):
  for x in range(128):
    r = first_img[x,y][0]
    g = first_img[x,y][1]
    b = first_img[x,y][2]

    copy_img[x][y][0] = truncate(r + d)
    copy_img[x][y][1] = truncate(g + d)
    copy_img[x][y][2] = truncate(b + d)

plt.imshow(copy_img)
plt.axis('off')
```

#Contraste

```python
first_img = df_flowers.loc[0]['image']
copy_img = first_img.copy()
# Converting image into a numpy array with a dimension (width, height, 3)
data = np.array(copy_img)

# Calculate average brightness
μ = np.mean(data, axis=2)
μ_mean = μ.mean()
width = 128
height = 128

beta = 100
# Calculate factor
if beta == 255: alpha = np.infty
else: alpha = (255+beta)/(255-beta)

for x in range(width):
    for y in range(height):
        r, g, b = copy_img[x][y]

        r_ = truncate(alpha*(r - μ_mean) + μ_mean)
        g_ = truncate(alpha*(g - μ_mean) + μ_mean)
        b_ = truncate(alpha*(b - μ_mean) + μ_mean)

        new_pixel = (int(r_), int(g_), int(b_))
        copy_img[x][y] = new_pixel

plt.subplot(1,2,1)
plt.imshow(copy_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(first_img)
plt.axis('off')
```

#Gamma

```python
first_img = df_flowers.loc[50]['image']
image = first_img.copy()
# Converting image into a numpy array with a dimension (width, height, 3)
data = np.array(image)

gamma = 3

width = 128
height = 128

for x in range(width):
    for y in range(height):
        r, g, b = image[x][y]

        r_ = 255 * (r/255)**gamma
        g_ = 255 * (g/255)**gamma
        b_ = 255 * (b/255)**gamma

        new_pixel = (int(r_), int(g_), int(b_))
        image[x][y] = new_pixel

plt.subplot(1,2,1)
plt.imshow(image)
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(first_img)
plt.axis('off')
```

#Blur

```python
first_img = df_flowers.loc[5]['image']
image = first_img.copy()

# Load image pixels
data = np.array(image)
kernel_size = 2
# Half kernel size
a = kernel_size // 2


width = 128
height = 128

for x in range(width):
    for y in range(height):
        # min and max are used for edge pixels
        x_start = max(x - a, 0)
        y_start = max(y - a, 0)

        x_end = min(x + a + 1, width)
        y_end = min(y + a + 1, height)

        # Array with all pixels inside square
        square = data[x_start:x_end, y_start:y_end]

        # We take their average
        r_, g_, b_ = square.mean(axis=(0,1))
        
        new_pixel = (int(r_),int(g_), int(b_))
        image[x][y] = new_pixel

plt.subplot(1,2,1)
plt.imshow(image)
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(first_img)
plt.axis('off')
```

```python
img = df_flowers.iloc[0]['image']

#M = np.array([0]*128*128*3).reshape(49152)
M = pd.DataFrame()
img = df_flowers.iloc[0]['image']
img2 = df_flowers.iloc[1]['image']
for z in range(210):
  img = df_flowers.iloc[z]['image']
  for y in range(128):
    for x in range(128):
      M.loc[z,128*y] = img[x][y][0]
      M.loc[z,(128*128) + (128*y) + x] = img[x][y][1]
      M.loc[z,(128*128*2) + (128*y) + x] = img[x][y][2]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=4)
principalComponents = pca.fit_transform(M.values)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2','principal component 3', 'principal component 4'])
```
